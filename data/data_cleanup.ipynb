# City Bike Data
All of the data came from this website: https://www.citibikenyc.com/system-data
Years 2018, 2019 and 2020 were used. Per the website, data is stored by month in zip files which contained a csv file for the data. Due to the size of the data (the data is going to be used in Tableau public, and there are limitations on how much data can be used), the scope was narrowed to the same two months for each year: June, and July. 
The data was imported into the notebook as a csv and read into a dataframe. The dataframes for each year were concatenated into one year. Those dataframes were written to csv files. 

```python
import pandas as pd
```

```python
bikes201806 = pd.read_csv("csv_files/input/201806-citibike-tripdata.csv")
bikes201806.head()
```

```python
bikes201807 = pd.read_csv("csv_files/input/201807-citibike-tripdata.csv")
bikes201807.head()
```

```python
bikes201907 = pd.read_csv("csv_files/input/201907-citibike-tripdata.csv")
bikes201907.head()
```

```python
bikes2018 = pd.concat(frames2018)
```
