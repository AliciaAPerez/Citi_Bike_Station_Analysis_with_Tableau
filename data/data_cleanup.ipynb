# City Bike Data
All of the data came from this website: https://www.citibikenyc.com/system-data
The data was pulled in from csv files into dataframes. Years 2019 and 2020 were used, and each had 1 file/dataframe per each month of the year. The dataframes were originally going to be concatenated into one year. However, due to the size of the data. They were grouped into quarts with three months each for each year making 8 dataframes. Those dataframes were written to csv files.
